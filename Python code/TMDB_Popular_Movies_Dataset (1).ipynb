```python
!pip install tmdbsimple
import csv
import tmdbsimple as tmdb

# Replace 'YOUR_API_KEY_HERE' with your actual TMDb API key
tmdb.API_KEY = 'c336e6f61c7e149ad9c456bdf451d8c7'

def get_popular_movies_keywords():
    # Number of pages to request
    num_pages = 100  # 20 pages * 20 movies per page = 400 movies

    # Initialize an empty list to store keywords
    all_keywords = []

    for page in range(1, num_pages + 1):
        # Discover popular movies for the current page
        discover = tmdb.Discover()
        popular_movies = discover.movie(sort_by='popularity.desc', language='en-US', page=page)

        # Extract keywords for each popular movie on the current page
        for movie in popular_movies['results']:
            movie_id = movie['id']
            movie_title = movie['title']

            # Get keywords for the current movie
            movie_keywords = get_movie_keywords(movie_id)

            # Add the movie ID, title, and keywords to the list
            all_keywords.append({
                'movie_id': movie_id,
                'movie_title': movie_title,
                'keywords': movie_keywords,
            })

    return all_keywords

def get_movie_keywords(movie_id):
    # Get keywords for the given movie ID
    movie = tmdb.Movies(movie_id)
    keywords = movie.keywords()

    # Extract keyword names
    keyword_names = [kw['name'] for kw in keywords['keywords']]

    return keyword_names

def write_to_csv(data, csv_filename='movie_keywords.csv'):
    # Open a CSV file for writing
    with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
        # Create a CSV writer
        csv_writer = csv.writer(csv_file)

        # Write the header row
        csv_writer.writerow(['Movie ID', 'Movie Title', 'Keywords'])

        # Write rows to the CSV file
        for entry in data:
            movie_id = entry['movie_id']
            movie_title = entry['movie_title']
            keywords = entry['keywords']
            csv_writer.writerow([movie_id, movie_title, ', '.join(keywords)])

def main():
    # Get keywords for the first 400 popular movies
    popular_movies_keywords = get_popular_movies_keywords()

    # Write the data to a CSV file
    write_to_csv(popular_movies_keywords)

if __name__ == "__main__":
    main()
```

```python
import tmdbsimple as tmdb
import csv

# Replace 'YOUR_API_KEY_HERE' with your actual TMDb API key
tmdb.API_KEY = 'c336e6f61c7e149ad9c456bdf451d8c7'

def get_movie_genres(movie_id):
    # Get genres for the given movie ID
    movie = tmdb.Movies(movie_id)
    genres = movie.info().get('genres', [])

    # Extract genre names and IDs
    genre_data = [{'id': genre['id'], 'name': genre['name']} for genre in genres]

    return genre_data

def get_popular_movies_genres():
    # Number of pages to request
    num_pages = 100  # 20 pages * 20 movies per page = 400 movies

    # Initialize an empty list to store genre data
    all_genres = []

    for page in range(1, num_pages + 1):
        # Discover popular movies for the current page
        discover = tmdb.Discover()
        popular_movies = discover.movie(sort_by='popularity.desc', language='en-US', page=page)

        # Extract genres for each popular movie on the current page
        for movie in popular_movies['results']:
            movie_id = movie['id']
            movie_title = movie['title']

            # Get genres for the current movie
            movie_genres = get_movie_genres(movie_id)

            # Add the movie title, genre names, and genre IDs to the list
            all_genres.append({
                'movie_title': movie_title,
                'genre_names': ', '.join([genre['name'] for genre in movie_genres]),
                'genre_ids': ', '.join([str(genre['id']) for genre in movie_genres])
            })

    return all_genres

def write_genres_to_csv(data, csv_file_path):
    # Write genre data to CSV file
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        fieldnames = ['movie_title', 'genre_names', 'genre_ids']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

        for entry in data:
            writer.writerow(entry)

def main():
    # Get genres (names and IDs) for the first 400 popular movies
    popular_movies_genres = get_popular_movies_genres()

    # Specify the CSV file path
    csv_file_path = 'movie_genres.csv'

    # Write genre data to CSV file
    write_genres_to_csv(popular_movies_genres, csv_file_path)
    print(f"CSV file created at: {csv_file_path}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import chardet
import io

# Replace '/content/keywords.csv' with the actual path to your CSV file
csv_file_path = '/content/keywords.csv'

# Detect the encoding using chardet
with open(csv_file_path, 'rb') as rawdata:
    result = chardet.detect(rawdata.read())

# Read the CSV file with the detected encoding
df = pd.read_csv(csv_file_path, encoding=result['encoding'])

# Function to clean non-UTF-8 compatible characters
def clean_non_utf8(value):
    if isinstance(value, str):
        return value.encode('utf-8', 'ignore').decode('utf-8')
    else:
        return value

# Apply the cleaning function to each cell
df = df.applymap(clean_non_utf8)

# Save the DataFrame to a new CSV file with UTF-8 encoding
utf8_csv_file_path = '/content/keywords_utf8.csv'
df.to_csv(utf8_csv_file_path, index=False, encoding='utf-8')

print(f"CSV file saved with UTF-8 encoding: {utf8_csv_file_path}")
```
